```python
import datetime

# List of date strings
date_strings = train['DATE (MM/DD)']

# Convert dates to a standard format
standard_dates = []
for date_str in date_strings:
    # Split date string into day and month components
    parts = date_str.split('-')
    day_str = parts[0]
    month_str = parts[1]
    
    format_str = "%d-%b"
    
    if len(day_str) == 2 and len(month_str) == 3:
        # "dd-mm" format
        date_str = f"{int(day_str):02d}-{month_str}"
    else:
        # "mm-dd" format
        date_str = f"{int(month_str):02d}-{day_str}"
    

    
    # Determine if day and month components are in "dd-mm" or "mm-dd" format
 
    if date_str.split("-")[1] == "Feb" and date_str.split("-")[0] == "29":
        standard_dates.append(date_str)
    else:
        # Convert date string to datetime object and then back to string in standard format
        date = datetime.datetime.strptime(date_str, format_str)
        standard_date = date.strftime('%m-%b')
        standard_dates.append(standard_date)

#print(standard_dates)
```

```python
def createdate(inp_str):
    day, month_str = inp_str.split("-")
    month_dict = {"Jan": "01", "Feb": "02", "Mar": "03", "Apr": "04", "May": "05", "Jun": "06",
              "Jul": "07", "Aug": "08", "Sep": "09", "Oct": "10", "Nov": "11", "Dec": "12"}
    month = month_dict[month_str]
    return f"{day.zfill(2)}{month}"
```

```python
import cv2
import numpy as np

def elfan(image):
    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Threshold the image to separate clouds from sky
    ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

    # Find contours of cloud pixels
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    # Calculate cloud cover percentage
    cloud_area = sum([cv2.contourArea(cnt) for cnt in contours])
    total_area = image.shape[0] * image.shape[1]
    percent_cloud_cover = cloud_area / total_area * 100
    
    return percent_cloud_cover
```

```python
image = cv2.imread("/kaggle/input/cloud-cover/CCD DATA SET-20230311T034557Z-001/CCD DATA SET/0101/0101091000.jpg")

# Call the elfan function to detect percent cloud cover
percent_cloud_cover = elfan(image)
print(percent_cloud_cover)
```
